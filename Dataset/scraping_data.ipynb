```python
import pandas as pd
pd.options.mode.chained_assignment = None
import csv
import numpy as np

from google_play_scraper import Sort, reviews, reviews_all
from itertools import count
```

```python
scrapreview, continual_token = reviews(
    'com.miHoYo.GenshinImpact',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=11000,
    filter_score_with=None
)
```

```python
df = pd.DataFrame(np.array(scrapreview),columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df
```

```python
df.info()
```

```python
new_df = df[['reviewId', 'userName', 'userImage', 'score', 'content', 'thumbsUpCount', 'at']]
new_df.info()
```

```python
clean_df = new_df.dropna()
clean_df = clean_df.drop_duplicates()
clean_df.info()
```

```python
clean_df.head()
```

```python
clean_df.to_csv('genshin_reviews.csv', index=False)
```
